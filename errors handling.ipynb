```python
# exceptions
# these happend while the program in running ,even if the syntax is correct
x = 10 / 0
# so here this code is syntacitally correct,but the operation is invalid
```

```python
# why do we use excepetions
# catch errors safely
# prevent our program from crashing
# shpw custom messages to the user
# continue running even after an error
# without exception handling the whole program would stop on first error
# syntax:
try:
    result = 10 / 0
except ZeroDivisionError:
    print("cannot divide a number by zero!")
```

```python
# handling mulitple exceptions
# one try block can raise different types of errors
# we handle multiple exceptions to
try:
    a = int(input("enter a number: "))
    b = int(input("enter another number: "))
    print(a * b)


except ZeroDivisionError:
    print("you cannort divide by zero")

except ValueError:
    print("please enter valid numbers")

    
```

```python
# try - except-else-finally with multiple exceptions
try: 
    num = int("10")
    print( "10" / num)
except ZeroDivisionError:
    print("we cannot divide")
except ValueError:
    print("conversion error")
else:
    print("no exception occurred")
finally:
    print("execution finished")
    
```
