```python
# iterator 
# an iterator is an object in python that we can loop through one item at a time
# an object is an iterator if it has two methods
# __iter__()
# __next__()
# whydo we use iterator 
# to go through items one by obe (like a list,tuples,string,etc)
# example of an iterator
nums = [1,2,3,4,5,6]
it = iter(nums)
print(it)
print(next(it))
print(next(it))
print(next(it))
```

```python
# create iterator
class count():
    def __init__(self):
        self.num = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.num <= 9:
            x = self.num
            self.num += 1
            return x
        else:
            raise StopIteration

c = count()

for i in c:
    print(i)
```

```python
# what is generator
# a generator is a simple way to create an iterator using functions
# and the keyword yeild
# example
def my_gen():
    yield 1
    yield 2
    yield 3
g = my_gen()
print(next(g))
print(next(g))
print(next(g))
```

```python
def squares(n):
    for i in range(1, n+1):
        yield i * i

for num in squares(5):
    print(num)
```

```python
# map
# it applies a function to nums every item in a list
# example:double each numbers
# using map
nums = [1,2,3,4]
result = list(map(lambda x : x * 2, nums)) 
print(result)
```

```python
# same example without using map (using loop)
nums = [1,2,3,4]
result = []
for x in nums:
    result.append(x + 3)

print(result)
```

```python
# filter()
# what filter does:
# it keeps the only condition that match a cindition(true/false)
# example:keep even numbers
# using filter
nums = [1,2,3,4,5,6,7,8]
result = list(filter(lambda x : x % 2 == 0, nums))
print(result)
```

```python
# same example without filter()
nums = [1,2,3,4,5,6,7,8]
result = []
for x in nums:
    if x % 2 == 0:
        result.append(x)
print(result)
                      
```

```python
# reduce()
# it reduces a list to a single value(sum,product,max,ect)
# example: sum of all numbers
from functools import reduce

nums = [1,2,3,4,5]
result = reduce(lambda a,b: a + b,nums)
print(result)
```

```python
# without reduce:
nums = [1,2,3,4,5]
total = 0
for x in nums:
    total += x

print(total)
```
