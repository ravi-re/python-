```python
# dict comprension
# dictionary comprension is a short and clean wY TO CREATE DICTIONARY using a single line od code
# {key : value for item in something}
# converting two list into a dict
keys = ["product" , "price"]
values = ["laptop",1200]
result= {k: v for k, v in zip(keys ,values)} 
print(result)
```

```python
# without comprension 
squares  = {}
for i in range(5):
    squares[i] = i ** 2 
print(squares)
```

```python
# with comprension
squares = {i: i * 2 for i in range(5)}
print(squares)
```

```python
# shallow copy > copies only the outer structure
# deep copy > copies everything (outer + inner abjects)
# if we change inner data > both changes 
# shallow copy
# change innner element
a  = [[1,2],[3,4]]
b = a.copy()

b[1][0] = 108      
print(a)
print(b)
```

```python
# deep copy
# copies the entire structure
# inner lists/dicts get new copies
# changes do not affect the orginal
import copy
d = copy.deepcopy(c)
c = [[4,5],[6,7]]
d = copy.deepcopy(c)

d[0][1] = 29
print(c)
print(d)
# only d changed
# deep copy create a completely independent object
```

```python
 # prime number function
def is_prime(n):
    if n <= 1:
        return False
    for i in range (2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
print(is_prime(2))
print(is_prime(7))
```

```python
# real-time example with input
num = int(input("enter a number: "))

if is_prime(num):
    print("prime number")
else:
    print("not a prime number")

```
