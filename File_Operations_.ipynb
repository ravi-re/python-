# File Operations in Python

## Working with Text Files

```python
# Writing to a text file
with open(r"C:\Users\kotar\Desktop\Amar\python\notes\sample.txt", "w") as f:
    f.write("Hello, this is a sample text file.\n")
    f.write("It has multiple lines.\n")
    f.write("hi hello world\n")
    f.write('hello this is python testing file operations commands')

# Reading from a text file
with open(r"C:\Users\kotar\Desktop\Amar\python\notes\sample.txt", "r") as f:
    content = f.read()
    print("File content:")
    print(content)
```

```python
# Reading line-by-line
with open(r"C:\Users\kotar\Desktop\Amar\python\notes\example.txt", "r") as f:
    for line in f:
        print("Line:", line.strip())
```

```python
with open("example.txt", "w") as f:
    f.write("Hello, this is a sample text file.\n")
    f.write("It has multiple lines.\n")
    f.write("hi hello world\n")
    f.write("am joey\n")
```

```python
with open("example.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
# Append
# Adds content at the end of file without erasing existing data.

with open("example.txt", "a") as file:
    file.write("Line 5: This is appended.\n")
    file.write("Line 6: This is python session 5\n")
```

```python
#first 10 charc
with open("example.txt", "r") as file:
    print("First 10 characters:", file.read())
```

```python
#read line by line
with open("example.txt", "r") as file:
    print("Line 1:", file.readline().strip())
    print("Line 2:", file.readline().strip())
    print("Line 3:", file.readline().strip())
```

```python
#into list
with open("example.txt", "r") as file:
    lines = file.readlines()
    print(lines)
    
```

```python
#copy file 
with open("example.txt", "r") as src, open("copy_sample.txt", "w") as dest:
    for line in src:
        dest.write(line)


with open(r"copy_sample.txt", "r") as f:
    content = f.read()
    print("File content:")
    print(content)
```

```python
with open("example.txt", "r+") as file:
    content = file.read()
    file.seek(35)
    file.write("chandler bing")
```

## Working with Excel Files (.xlsx)

```python
import pandas as pd

# Sample data
data = {
    "Name": ["Alice", "Bob", "Charlie", 'ruchik','sushma','Thauja'],
    "Marks": [85, 90, 78,87,11,44]
}
df = pd.DataFrame(data)

df
# Writing to an Excel file
df.to_excel("students.xlsx", index=False)

# # Reading from an Excel file
df_read = pd.read_excel("students.xlsx")
print("Data from Excel:")
print(df_read)
```
