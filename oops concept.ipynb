```python
# inheritance 
# means one class (child) gets theb properties and behavior of other class(parent)
# ot helps in code reusability and creating a logicaal hierarchy
# example: a child inherits features from parent(name,behavior),etc)
# basic inheritance
# rxample:vehicle --> car
class vehicle:
    def start(self):
        print("vehicle is starting")

class car(vehicle):
    def drive(self):
        print("car is driving")
my_car = car()
my_car.start()
my_car.drive()
```

```python
# types of inheritance
# single inheritance
# one parent ----> one child
class animal:
    def eat(self):
        print("animal eats")

class Dog(animal):
    def bark(self):
        print("dogs barks")

dog = Dog()
dog.eat()
dog.bark()
```

```python
# multilevel inheritance
# parent --> child ---> grandchild
class grandparents:
    def house(self):
        print("has a house")
class parent(grandparents):
    def car(self):
        print("has a car")
class child(parent):
    def bike(self):
        print("has a bike")

c = child()
c.house()
c.car()
c.bike()
    
```

```python
# hierarchione inheritance
# one parent ----> multiple children
class shape:
    def area(self):
        print("calculating area")

class circle(shape):
    pass

class square(shape):
    pass

circle().area()
square().area()
```

```python
# method overriding
# child class redefines parents's method
class vehicle:
    def start(self):
        print("vehicle starts")
class bike(vehicle):
    def start(self):
        print("bike starts with button")
b = bike()
b.start()
```

```python
# using super()
# call parent class method inside child class
class vehicle:
    def start(self):
        print("vehicle started")
class car(vehicle):
    def start(self):
        super().start()
        print("car ready to drive")
car().start()

# how the start method works here is 
# method overriding
# both vehicle and car have a method named start()
# car overriding the start() method of vehicle
# ---------------------------------------------------------------------
# what happens when we call the method
# c = car()
# c.start()
# python looks for start() in car
# it finds car.start(),so it runs that method
# inside car.start(),super().start() is called
# ------------------------------------------------------------------------
# what super().start()does
# super()refers to the parent class(vehicle)
# it calls vehicle.start()
# this prints
# vehicle started
# -----------------------------------------------------------------------
# after parent method runs
# control comes back to car.start()
# them this line runs:
# print("car ready to drive")
# -----------------------------------------------------------------------
# the child method runs first,calls the parent method
# using super(),and then adds its own behavior
```

```python
# abstraction = showingb essential features and hiding unnecessary details
# python provides abstraction using (ABC) abstract base classes
# example: payment system (real-world)
from abc import ABC, abstractmethod

class payment(ABC):
    @abstractmethod
    def make_payment(self,amount):
        pass

class creditcardpayment(payment):
    def make_payment(self, amount):
        print(f"paid $ {amount} using credit card")

class upipayment(payment):
    def make_payment(self, amount):
        print(f"paid ${amount} using upi")
payment = upipayment()
payment.make_payment(500) 
```

```python
# encapsulation
# wrapping data(variables) and methods into a single unit(class)
# and controlling access to the data to protect 
# encapsluation = data hiding + contorlled access
# example: we cannot directly change our ban balance 
# we must use:
# deposit()
# withdraw()
# here balance is protected
class Bankaccount:
    def __init__(self, balance):
        self.__balance = balance                                     # private variable

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("insufficent balance")

    def get_balance(self):
        return self.__balance



account = Bankaccount(500)

account.deposit(1000)
print(account.get_balance())
```

```python
# polymorphism means one name many forms 
# in python the same methofd or function nhaves differently depending on the object that is calling it
# real-time example
# a play button:
# on youtube ---> plays video
# on music app --->  plays music
# same actions,different behavior ---->polymorphims
# using the same method name
class dog:
    def speak(self):
        print("dog barks")
class cat:
    def speak(self):
        print("cat meows")

animals = [dog(), cat()]
for animal in animals:
    animal.speak()
```

```python
# agrs* 
# it means allows a function to accept any number of positional arguments
# arguments are stored as a tuple
def add_numbers(*args):
    total = 0
    for num in args:
        total += num
    print(total)

add_numbers(10,20,30)
```

```python
# **kwargs(keyword argument)
# it means allows a function to accept any number of keyword arguments
# arguments are stored in a dictionary 
def show_details(**kwargs):
    for k,v in kwargs.items():
        print(k,":",v)

show_details(name = "ravi", age=24,city = "new york")
```

```python
# using *arge and **kwargs together
def student_info(*args,**kwargs):
    print("subjects:", args)
    print("details:",kwargs)

student_info("math","science", name = "ravi", grade ="A") 
```

```python
# decorator
# a decorator is a function that:
# adds extra behavior to another function without changing its code
# example:
# think of a gift wrraping:
# gift = orginal function
# wrapper = decorator
# gift remains same but extra layer is added
# example:
def my_decorator(func):
    def wrapper():
        print("before function runs")
        func()
        print("after function runs")
    return wrapper


@my_decorator
def say_hello():
    print("what is this!")

say_hello()
```
